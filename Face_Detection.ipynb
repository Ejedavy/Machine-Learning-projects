```python
import cv2 as cv
import numpy as np
```

```python
path = input("Enter the absolute path to the image: ")
```

```python
path
```

```python
image = cv.imread(path)
```

```python
gray = cv.cvtColor(image, cv.COLOR_BGR2RGB)
```

```python
cascade = cv.CascadeClassifier("/content/haarscade.xml")
```

```python
faces = cascade.detectMultiScale(gray, scaleFactor = 1.1, minNeighbors = 3)
```

```python
for (x,y,w,h) in faces:
  cv.rectangle(image, (x,y), (x+w, y+h),(0,255,0),2)
```
